```python
# Print Pattern

# Method 1:
print("* *****")
print("** ****")
print("*** ***")
print("**** **")
print("***** *")

# Method 2:
print("""
* *****
** ****
*** ***
**** **
***** *
""")

# Method 3:
print("* *****", "** ****", "*** ***", "**** **", "***** *", sep="\n")

# Method 4:
for i in range(1, 6):
    print("*" * i + " " + "*" * (6 - i))
```
