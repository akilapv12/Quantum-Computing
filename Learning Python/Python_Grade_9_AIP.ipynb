```python
# Print Pattern

# Method 1:
print("* *****")
print("** ****")
print("*** ***")
print("**** **")
print("***** *")

# Method 2:
print("""
* *****
** ****
*** ***
**** **
***** *
""")

# Method 3:
print("* *****", "** ****", "*** ***", "**** **", "***** *", sep="\n")

# Method 4:
for i in range(1, 6):
    print("*" * i + " " + "*" * (6 - i))
```

```python
# Find Square Root:


num = input("Enter a number : ")
import math
print("Square Root of Num: ", math.sqrt(float(num)))
```

```python
# Addition

num1 = input("Enter first number : ")
num2 = input("Enter second number : ")
print(num1, "+", num2, "=", float(num1) + float(num2))
```

```python
# km ---> m

km = input("Enter distance in km: ")
m = float(km)*1000
print(km, "km = ", m, "m")
```

```python
# Table of 5

x = int(input("Enter how many terms of 5's table you want: "))
i = 1
while i<=x:
  print(x,"*",i,"=",x*i)
  i += 1
```

```python
# Simple Intrest

p = float(input("Enter principal amount: "))
r = float(input("Enter rate of intrest: "))
t = float(input("Enter time (years): "))
si = (p*r*t)/100
print("Simple Intrest: ", si)
```

```python
# Area and Perimeter of Rectangle:

l = float(input("Enter length: "))
b = float(input("Enter breadth: "))
print("\nArea: ", l*b)
print("Perimeter: ", 2*(l+b))
```

```python
# Area of Triangle:

b = float(input("Enter base: "))
h = float(input("Enter height: "))
print("\nArea: ", 0.5*b*h)
```
