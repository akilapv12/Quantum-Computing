```python
# Print Pattern

# Method 1:
print("* *****")
print("** ****")
print("*** ***")
print("**** **")
print("***** *")

# Method 2:
print("""
* *****
** ****
*** ***
**** **
***** *
""")

# Method 3:
print("* *****", "** ****", "*** ***", "**** **", "***** *", sep="\n")

# Method 4:
for i in range(1, 6):
    print("*" * i + " " + "*" * (6 - i))
```

```python
# TSA, LSA, and Vol. of Cuboid

l = float(input("Enter length: "))
b = float(input("Enter breadth: "))
h = float(input("Enter height: "))
print("\nTotal Surface Area = ", 2*(l*b+b*h+l*h))
print("Lateral Surface Area = ", 2*h*(l+b))
print("Volume = ", l*b*h)
```

```python
# +ve, -ve, or 0
num = float(input("Enter a number: "))
if num>0:
  print(f"The number {num} is possitive.")
elif num==0:
  print(f"The number {num} is zero.")
else:
  print(f"The number {num} is negative")
```

```python
# Natural Numbers
times = int(input("Enter how many times you want to print: "))
i = 1
while i<=times:
  print(i)
  i += 1
```

```python
# Even Numbers
times = int(input("Enter how many times you want to print: "))
i = 2
while i<=times*2:
  print(i)
  i += 2
```

```python
# Odd numbers
times = int(input("Enter how many times you want to print: "))
i = 1
while i<=times*2:
  print(i)
  i += 2
```

```python
# Sum of Nums
times = int(input("Enter how many times you want to print: "))
i = 1
total = 0
while i<=times:
  total += i
  i += 1
print(total)
```
