```python
#Triangle Type
a=int(input("Enter the first side: "))
b=int(input("Enter the second side: "))
c=int(input("Enter the third side: "))
if a==b==c:
  print("Equilateral Triangle")
elif (a==b and b!=c) or (b==c and c!=a) or (c==a and a!=b):
   print("Isosceles Triangle")
else:
  print("Scalene Triangle")
```

```python
# ATM Simulator
pin = 31312
balance = 300

print("Welcome User!\n")
PIN = int(input("Enter your PIN: "))

if PIN == pin:
  print("Access Granted.\n")
  while True:
   print("1 -----> Check Balance")
   print("2 -----> Withdraw Money")
   print("3 -----> Deposit Money")
   print("4 -----> Exit\n")

   user = int(input("Enter your choice (Digits 1-4): "))
   print()

   if user == 1:
     print("Your balance is", balance, "AED.\n")

   elif user == 2:
     while True:
       withdrawal = int(input("How much AED would you like to withdraw: "))
       if withdrawal <= balance:
         balance -= withdrawal
         print("Please take your cash.")
         print("Your new balance is", balance, "AED.\n")
         break
       else:
         print("Insufficient Funds. Try a smaller amount.\n")

   elif user == 3:
     deposit = int(input("How much AED would you like to deposit: "))
     balance += deposit
     print("Your money has been deposited.")
     print("Your new balance is", balance, "AED.\n")

   elif user == 4:
     print("Thank you. Have a great day!")
     break

   else:
     print("Invalid choice. Please select a number between 1 and 4.\n")

     print("1 -----> Yes")
     print("2 -----> No")
     user2 = int(input("Would you like to proceed with other choices? "))
     print()

     if user2 != 1:
       print("Thank you for using the ATM!")
       break

else:
    print("Access Denied.")
```
